```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from scikeras.wrappers import KerasClassifier
import pickle
```

```python
df_full =pd.read_csv("./data/processed_data.csv")
```

# Split data

```python
df = df_full.copy()
Y_COL = "toxic"
# Target variable -> comment was upvoted

Y = df[Y_COL] 
del df[Y_COL]
X = df

X_full_train, X_test, Y_full_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)

X_train, X_val, Y_train, Y_val  = train_test_split(X_full_train, Y_full_train, test_size=0.25, random_state=1)

```

```python
# Build the neural network model with Input layer
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compile the model
# model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['auc'])
# history = model.fit(X_full_train, Y_full_train, epochs=5, batch_size=32, validation_split=0.2)
```

```python
wrapped_model = KerasClassifier(model=model, epochs=4, batch_size=32, validation_split=0.2, optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['auc'])
wrapped_model.fit(X_full_train, Y_full_train)
```

```python
from sklearn.metrics import roc_auc_score

Y_pred = wrapped_model.predict(X_test)
auc_score = roc_auc_score(Y_test, Y_pred)
print(f"AUC Score: {auc_score}")
```

```python
with open('./data/tfidf_vectorizer.pkl', 'rb') as f:
    tfidf_vectorizer = pickle.load(f)
```

```python
# Create the pipeline using the fitted vectorizer and the wrapped model
inference_pipeline = Pipeline([
    ('vectorizer', tfidf_vectorizer),  # Use the already-fitted vectorizer
    ('classifier', wrapped_model) # Use the wrapper with the model builder
])

with open('./data/text_classification_pipeline.pkl', 'wb') as f:
    pickle.dump(inference_pipeline, f)
```
